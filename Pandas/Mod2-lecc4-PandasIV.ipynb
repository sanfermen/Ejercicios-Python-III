8. Si extraemos los valores únicos de la columna species nos podemos dar cuenta que está un poco sucia y extraer informacion útil de la misma se puede convertir en algo difícil. En este ejercicio tendréis que extraer los ataques que han ocurrido por tiburón blanco (lo encontraremos en la columna species como algo que contenga white o White). Usad el método str.contains() para resolver este ejercicio.

```python
# Buscamos los valores únicos de la columna especies para conocer todos los tipos de tiburón
tiburon["species_"].unique()
```

```python
# Usando regex, buscamos los ataques producidos por tiburones blancos
patron = "[w|W]hite"
blanco = tiburon[tiburon["species_"].str.contains(patron, regex= True)]
blanco.head(2)
```

```python
# comprobamos las categorías que hay ahora en species
blanco["species_"].unique()
```

9. Extraed ahora los ataques por tiburón blanco o tigre. Utilizad el método str.contains( ) para resolver este ejercicio.

```python
# Usando regex, buscamos los ataques producidos por tiburón blanco o tigre
patron2 = "[w|W]hite|[t|T]iger"
blanco_tigre = tiburon[tiburon["species_"].str.contains(patron2, regex= True)]
blanco_tigre.head()
```

```python
# Comprobamos las especies diferentes
blanco_tigre["species_"].unique()
```

10. Utilizando la columna de year cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI. Donde el año sea menor que 2000 pondremos siglo-XX y el resto de los años siglo-XXI.

```python
# Creamos una nueva columna, que nos indique el siglo según el año en el que se ha producido el ataque
condiciones = [tiburon["year"] <2000, tiburon["year"]>=2000]
opciones = ["siglo-XX", "siglo-XXI"]
tiburon["siglo"] = np.select(condiciones,opciones)
tiburon.head()
```
