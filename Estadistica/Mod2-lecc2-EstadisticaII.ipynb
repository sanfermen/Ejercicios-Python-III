1.Extraer los siguentes estadísiticos del conjunto de datos para varias de las varibles disponibles:

1.1. Utilizando métodos de pandas: media, mediana, moda, máximo y mínimo

```python
# Media
print("Media del precio de la leche es:", df["Milk"].mean())
print("Media del precio de una entrada de cine:", df["Cinema"].mean())
print("Media del precio de la escuela primaria:", df["Primary School"].mean())
```

Con estos datos podemos afirmar que el precio medio de la leche entre todas las ciudades analizadas seria 1.13$; de una entrada de cine 8.33$ y el precio medio de la escuela primaria unos 8684.30 $. La media no nos aportaria mucha mas informacion

```python
#Mediana
print("Mediana del precio de la leche:", df["Milk"].median())
print("Mediana del precio de una entrada de cine:", df["Cinema"].median())
print("Mediana del precio de la escuela primaria:", df["Primary School"].median())
```

Con la mediana sabemos el valor que tenemos en la posicion central de todos nuestros valores una vez ordenados, aunque tengamos muchos valores extremos nos daria informacion mas exacta del valor medio de los productos. En este caso la mediana tanto de la leche como de la entrada de cine se aproximan bastante al valor de la media que hemos visto antes pero en el caso de la escuela primaria estos dos datos se alejan, lo que podriamos deducir que tenemos muchos valores extremos en esta columna.

```python
#Moda
print("Moda del precio de la leche:", df["Milk"].mode()[0])
print("Moda del precio de una entrada de cine:", df["Cinema"].mode()[0])
print("Moda del precio de la escuela primaria:", df["Primary School"].mode()[0])
```

La moda nos va a dar el valor que mas se repite, en este caso hemos escogido el que se encuentra en posicion 0. En la columna de leche seria de 1.04$, la entrada de cine seria 12$ y la escuela primaria 20000$. Son los valores que han aparecido en mas ciudades.

```python
#Maximo
print("Maximo del precio de la leche:", df["Milk"].max())
print("Maximo del precio de una entrada de cine:", df["Cinema"].max())
print("Maximo del precio de la escuela primaria:", df["Primary School"].max())
```

El maximo nos da el valor mas alto por cada columna, en este caso en la columna de leche el precio mas alto seria de 6.81$, 39.92$ para la entrada de cine y 75000$ la escuela primaria. Nos podriamos hacer una idea del grado de dispersion que tienen los datos en esas columnas; por ejemplo en el caso de la escuela primaria la diferencia entre el maximo y la mediana nos viene a decir que tenemos valores muy altos y muy alejados del cuartil 50 (6280$)

```python
#Minimo
print("Minimo del precio de la leche:", df["Milk"].min())
print("Minimo del precio de una entrada de cine:", df["Cinema"].min())
print("Minimo del precio de la escuela primaria:", df["Primary School"].min())
```

Algo parecido nos pasa con el minimo, tambien nos sirve para hacernos una idea de lo dispersos que tenemos los valores en nuestra columna. Como vemos, el mas alejado de la mediana (del 50% de nuestros valores) sigue siendo la escuela primaria

1.2. Percentiles 25, 75

```python
#Calculamos los percentiles
q1_m, q3_m = np.nanpercentile(df["Milk"], [25,75]) 
print(f"Q1 de la columna Milk es igual a: {q1_m}")
print(f"Q3 de la columna Milk es igual a: {q3_m}")
```

En la columna de "Milk" el 25% de nuestros valores son iguales o estan por debajo de 0.82 y el 75% de los valores estaran mas debajo o seran igual a 1.3

```python
q1_c, q3_c = np.nanpercentile(df["Cinema"], [25,75]) 
print(f"Q1 de la columna Cinema es igual a: {q1_c}")
print(f"Q3 de la columna Cinema es igual a: {q3_c}")
```

En la columna de "Cinema" el 25% de nuestros valores son iguales o estan por debajo de 4.89 y el 75% de los valores estaran mas debajo o seran igual a 11.75$

```python
q1_p, q3_p = np.nanpercentile(df["Primary School"], [25,75]) 
print(f"Q1 de la columna Primary School es igual a: {q1_p}")
print(f"Q3 de la columna Primary School es igual a: {q3_p}")
```

En la columna de "Primary School" el 25% de nuestros valores son iguales o estan por debajo de 3243.04$ y el 75% de los valores estaran mas debajo o seran igual a 11931.53$

1.3. Rango intercuartilico

```python
iqr_m = q3_m - q1_m
print(f"El rango intercuartilico de la leche es: {iqr_m}")
```

En la columna "Milk" apenas hay un 0.48 de diferencia entre la Q3 y la Q1, lo que nos dice que la mayoria de nuestros valores estan muy cerca unos de otros. Los valores en esta columna son bastante homogeneos

```python
iqr_c = q3_c - q1_c
print(f"El rango intercuartilico de una entrada de cine es: {iqr_c}")
```

En la columna "Cinema" hay 6.86 de diferencia entre la Q3 y la Q1, entre el cuartil 25 y el 75 hay 6.86 de distancia

```python
iqr_p = q3_p - q1_p
print(f"El rango intercuartilico de la escuela primaria es: {iqr_p}")
```

En este caso el rango intercuartilico es muy alto. La diferencia entre cuartiles es mucho mayor, lo que nos puede decir que no tenemos unos datos homogeneos

1.5. Tablas de frecuencias

1.5.1. Variables cualitativas

```python
#Analizamos la frecuencia de la columna "city" agrupandola por "country"
df_pais = df.groupby("country").count().reset_index() 
df_pais.sort_values(by = "city", ascending = False)
```

Asi vemos que el pais con mas ciudades analizadas es EEUU seguido por Italia, India, Brasil y Reino Unido. Por abajo se ve que hay muchos paises en los que solo se ha analizado una unica ciudad.

1.5.2. Variables cuantitativas

```python
#Analizamos cuantas veces se repite un mismo valor en la columna "potato" 
df_patata = df.groupby("Potato").count().reset_index() 
df_patata.sort_values(by = "city", ascending = False)

```

El precio que mas se repite en la columna de Patatas es 1.04 obtenido en 88 ciudades diferentes

1.5.3. Frecuencia acumulada

```python
df_sum = df_patata["city"].cumsum()
df_sum
```

La frecuencia absoluta acumulada de la columna Potato es 4444 (La suma de las frecuencias absolutas por todas las ciudades)

```python
df_sum2 = df_pais["city"].cumsum()
df_sum2
```

La frecuencia absoluta acumulada de la columna Pais es 4874 (La suma de las frecuencias absolutas por todas las ciudades)

Tabla de frecuencias relativas

```python
frecuencia_relativa = (df_patata / df.shape[0]) *100
frecuencia_relativa.sort_values(by = "city", ascending = False)
```
