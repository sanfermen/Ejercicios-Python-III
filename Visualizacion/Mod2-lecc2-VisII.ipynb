```python
#Importamos las librerias
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
#Cargamos el dataset
df_tiburon = pd.read_csv("files/tiburon.csv")
df_tiburon.head(2)
```

1. Ayer creamos un par de gráficas para USA y España, donde visualizamos el número de ataques en cada país a lo largo de los años y el tipo de los ataques. Hoy tendréis que hacer dos gráficas con subplots:

- En la primera debéis mostrar los tipos ataques en USA y España, cada subplot corresponderá a un país. Debéis:
    - Poner nombre a los ejes de cada uno de los subplots
    - Poner un títulos a cada uno de los subplots
    - Poner el color de las líneas en negro.
    - Establecer el tamaño de la gráfica a 15 x 5
    - Ponerle marcadores a las gráficas
    - No incluir leyenda


```python
# Creamos un dataframe con los ataques sólo de USA despues del 2000
df_tiburon_USA_2000 = df_tiburon[(df_tiburon["country"] == "USA") & (df_tiburon["year"] >= 2000)]
# Hacemos un DF con la suma de los ataques por tipo
dfy = pd.DataFrame(df_tiburon_USA_2000.type.value_counts())
dfy.reset_index(inplace=True)
```

```python
# Creamos un dataframe con los ataques sólo en España despues del 2000
df_tiburon_SPAIN_2000 = df_tiburon[(df_tiburon["country"]=="SPAIN") & (df_tiburon["year"]>=2000)]
# Hacemos un DF con la suma de los ataques por tipo
dfy_Spain = pd.DataFrame(df_tiburon_SPAIN_2000.type.value_counts())
dfy_Spain.reset_index(inplace=True)
```

```python
# Creamos una figura que tenga 2 gráficas en 1 fila y 2 columnas, de tamaño 15, 5
fig,ax = plt.subplots(1,2, figsize = (15,5))

# Creamos dos gráficos de línea con color de línea negro y marcadores en círculo
ax[0].plot(dfy["index"], dfy["type"], color= "black", marker = "o", markersize = 8)
ax[1].plot(dfy_Spain["index"], dfy_Spain["type"], color= "black", marker = "o", markersize = 8)

# Ponemos titulo a cada subplot
ax[0].set_title("Ataques Tiburón USA")
ax[1].set_title("Ataques Tiburón SPAIN")

# Ponemos titulo a los ejes
ax[0].set_xlabel("Tipo de ataque")
ax[0].set_ylabel("Número de ataques")
ax[1].set_xlabel("Tipo de ataque")
ax[1].set_ylabel("Número de ataques")

plt.show()
```

- En la segunda debéis mostrar los ataques por año para cada uno de los países.
    - De nuevo tendréis que ponerle nombre a los ejes y título a cada gráfica.
    - En este caso el subplot deberá tener dos filas y una columna
    - No incluir leyenda
    - Cambiar el color de las barras

```python
# Creamos un DF con los ataques por año en USA
dfy_year = pd.DataFrame(df_tiburon_USA_2000.year.value_counts())
dfy_year.reset_index(inplace=True)
```

```python
dfy_year.head(2)
```

```python
# Creamos un DF con los ataques por año en Spain
dfy_year2 = pd.DataFrame(df_tiburon_SPAIN_2000.year.value_counts())
dfy_year2.reset_index(inplace=True)
```

```python
# Creamos una figura que tenga 2 gráficos en 1 columna y 2 filas, de tamaño 10x15
fig, ax = plt.subplots(2,1, figsize = (10,15))

# Creamos un gráfico de barras, con las barras de distinto color en cada una
ax[0].bar(dfy_year["index"], dfy_year["year"], color = "green")
ax[1].bar(dfy_year2["index"], dfy_year2["year"], color = "red")

# Ponemos título a las 2 gráficas
ax[0].set_title("Ataques Tiburón USA")
ax[1].set_title("Ataques Tiburón SPAIN")

# Ponemos títulos a los ejes de las gráficas
ax[0].set_xlabel("Años de ataque")
ax[0].set_ylabel("Número de ataques")
ax[1].set_xlabel("Años de ataque")
ax[1].set_ylabel("Número de ataques")

plt.show()
```

2. Cread un único gráfico donde juntéis las gráficas de los tipos de ataques en una sola. Debéis:
- Poner leyenda a la gráfica
- Cada línea debe ir en un color
- Ponerle nombre a los ejes y a la gráfica.
- Poner marcadores a las lineas.

```python
# Creamos una figura que tenga los datos de 2 DF en una sola
fig, ax = plt.subplots(figsize= (10,5))

# Creamos las dos gráficas, una con la línea azul para USA y otra roja para Spain. Personalizamos los marcadores
rects1 = ax.plot(dfy["index"], dfy["type"], color= "blue", marker = "<", markersize = 8)
rects2 = ax.plot(dfy_Spain["index"], dfy_Spain["type"], color= "red", marker = ">", markersize = 8)

# Ponemos título a la gráfica
ax.set_title("Comparación ataques Tiburón USA - SPAIN")

# Ponemos título a los ejes
ax.set_xlabel("Tipo de ataque")
ax.set_ylabel("Número de ataques")

# Ponemos leyenda a la gráfica
ax.legend(labels = ["Ataques USA", "Ataques Spain"])

plt.show()
```
